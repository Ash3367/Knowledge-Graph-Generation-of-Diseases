```python
# import pandas as pd
url = "https://en.wikipedia.org/wiki/COVID-19"
my_table = pd.read_html(url)
my_table[0]
```

```python

mal=[]
a = "Coronavirus disease 2019(COVID-19)"
b = "Coronavirus disease 2019(COVID-19).1"
for i in range(4,len(my_table[0][a])-2):
    symptoms = my_table[0][b][i].split(",")
#     print(symptoms)
    for s in symptoms:
        temp=[a,my_table[0][a][i],s]
        mal.append(temp)
print(mal)
print(type(mal))
```

```python

df = pd.DataFrame(mal)
writer = pd.ExcelWriter('Coronavirus disease 2019(COVID-19).xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='welcome', index=False)
writer.save()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

g = nx.Graph()
plt.figure(figsize=(15,15))
pos = nx.spring_layout(g, k=5)
for (die,cat,data) in mal:
    g.add_edge(die,cat)
    g.add_edge(cat,data)
nx.draw(g, with_labels = True, node_size = 3000, node_color="yellow")
plt.savefig("Coronavirus disease 2019(COVID-19).JPEG")
```
